```python
# Checking null values
cpa.isnull().sum()
```

```python
# Filling null value with mean
cpa['Income'].fillna(cpa['Income'].mean(), inplace=True)
```

```python
# Convert date column
cpa['Dt_Customer'] = pd.to_datetime(cpa['Dt_Customer'], dayfirst=True)
```

```python
# Convert object columns to category
cpa['Education'] = cpa['Education'].astype('category')
cpa['Marital_Status'] = cpa['Marital_Status'].astype('category')
```
